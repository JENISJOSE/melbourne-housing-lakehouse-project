```python
from pyspark.sql.functions import col

bronze_df = spark.table("default.bronze_melb_housing")

# Data quality checks
null_price_count = bronze_df.filter(col("Price").isNull()).count()
null_suburb_count = bronze_df.filter(col("Suburb").isNull()).count()

print(f"Null Price Records: {null_price_count}")
print(f"Null Suburb Records: {null_suburb_count}")

# Data cleaning
silver_df = (
    bronze_df
        .filter(col("Price").isNotNull())
        .filter(col("Suburb").isNotNull())
        .dropDuplicates()
)

print(f"Silver Row Count: {silver_df.count()}")

# Save Silver Layer
silver_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("default.silver_melb_housing")
```
